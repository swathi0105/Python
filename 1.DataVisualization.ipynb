#####Using Matplotlib: Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy

```python
# Loading matplotlib package
import matplotlib.pyplot as plt
%matplotlib inline
```

Rank of major vs payment

```python
df.plot(x="Rank", y=["P25th", "Median", "P75th"])
plt.show()
```

Observations

  

*   The median income decreases as rank decreases. This is expected because the rank is determined by the median income
*   Some majors have large gaps between the 25th and 75th percentiles. People with these degrees may earn significantly less or significantly more than the median income
*   Other majors have very small gaps between the 25th and 75th percentiles. People with these degrees earn salaries very close to the median income

Adding new column based on another column (data modification)

```python
conditions = [
    (df['total'] <= 10),
    (df['total'] > 10) & (df['total'] <= 15),
    (df['total'] > 15) & (df['total'] <= 20),
    (df['total'] > 20)
    ]

# create a list of the values we want to assign for each condition
values = ['Low', 'Moderate', 'High', 'Very high']

# create a new column and use np.select to assign values to it using our lists as arguments
df['type'] = np.select(conditions, values)

# display updated DataFrame
df.head()
```

```python
conditions = [
    (df['ins_premium'] <= 1000),
    (df['ins_premium'] > 1000)
    ]

# create a list of the values we want to assign for each condition
values = ['A', 'B']

# create a new column and use np.select to assign values to it using our lists as arguments
df['prem_type'] = np.select(conditions, values)

# display updated DataFrame
df.head()
```

```python
sns.relplot(data=df, x="speeding", y="alcohol", hue="type")
```

```python
sns.relplot(data=df, x="speeding", y="alcohol", hue="prem_type", col="type", col_wrap=2)
```

Visualisation

```python
#matplotlib and seaborn
#df.speeding.plot(kind="hist")
sns.distplot(df['speeding'],kde = True, bins = 20)
```

```python
#Comparison between different groups
sns.barplot(x = 'prem_type', y = 'alcohol', hue = 'type', data = df,
            palette = 'PiYG', #PiYG, GnBu, deep
            order = ['A', 'B'],  
            capsize = 0.05,             
            saturation = 8,             
            errcolor = 'gray', errwidth = 2,  
            ci = 'sd'
            )
#plt.legend().set_title('')
plt.show()
```

```python
#Violin plot
sns.violinplot(x = 'prem_type', y = 'alcohol', hue = 'type', data = df,
            palette = 'PuRd')
```

```python
#Point plot
sns.pointplot(x="type", y="alcohol", data=df)
```

```python
#Joint plot
#sns.jointplot(x="speeding", y="alcohol", data=df)
sns.jointplot(x="speeding", y="alcohol", data=df, kind='kde') #kde, hex
```

```python
#Regression plot (Linear)
ax = sns.regplot(x="speeding", y="alcohol", data=df)
```

Correlation analysis

```python
corr=df.corr()
print(corr)
sns.heatmap(corr,annot=False,linewidths=.5,cmap="YlGnBu")
```

```python
sns.set_style("ticks")
sns.pairplot(df,hue = 'type',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```

#####Using Plotly: Plotly provides online graphing, analytics, and statistics tools for individuals and collaboration, as well as scientific graphing libraries for Python, R, MATLAB

```python
import plotly as px
from plotly.offline import init_notebook_mode,iplot
import plotly.graph_objects as go
import cufflinks as cf
init_notebook_mode(connected=True)
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
#labels
lab = df["survived"].value_counts().keys().tolist()
#values
val = df["survived"].value_counts().values.tolist()
trace = go.Pie(labels=lab, 
                values=val, 
                marker=dict(colors=['red']), 
                # Seting values to 
                hoverinfo="value"
              )
data = [trace]
```

```python
layout = go.Layout(title="Survived Distribution")
fig = go.Figure(data = data,layout = layout)
fig.show(renderer="colab")
```

```python
# defining data
trace = go.Histogram(x=df['age'],nbinsx=40,histnorm='percent')
data = [trace]
# defining layout
layout = go.Layout(title="Age Distribution")
# defining figure and plotting
fig = go.Figure(data = data,layout = layout)
fig.show(renderer="colab")
```

```python
#defining data
trace = go.Scatter(x = df['age'],y=df['fare'],text = df['survived'],mode='markers')
data=[trace]
#defining layout
layout = go.Layout(title='Fare Vs Age Scatter Plot',xaxis=dict(title='Age'),yaxis=dict(title='Fare'),hovermode='closest')
#defining figure and plotting
figure = go.Figure(data=data,layout=layout)
figure.show(renderer="colab")
```

```python
import plotly.figure_factory as ff
```

```python
#defining data
a = df[df['pclass']==1]['fare']
b = df[df['pclass']==2]['fare']
c = df[df['pclass']==3]['fare']
hist_data=[a,b,c]
group_labels=['1','2','3']
#defining fig and plotting
fig = ff.create_distplot(hist_data,group_labels,bin_size=
[1,1,1],show_curve=False)
fig.update_layout(title_text='Distribution for Fares')
#iplot(fig)
fig.show(renderer="colab")
```

```python
#defining data
data=[go.Scatter(x = df['age'],y=df['fare'],
               text=df['pclass'],
                mode='markers',
               marker=dict(size=df['pclass']*15, color=df['survived'],showscale=True),
              )]
#defining layout
layout = go.Layout(title='Fare vs Age with Survivability and Pclass',xaxis=dict(title='Age'),yaxis=dict(title='Fare'),hovermode='closest')
#defining figure and plotting
figure = go.Figure(data=data,layout=layout)
figure.show(renderer="colab")
```
